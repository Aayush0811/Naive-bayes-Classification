```python
import numpy as np
```

```python
def normal_likelihood(training_features):
    
    training_features = np.array(training_features)
    
    mu_cap = np.mean(training_features)
    
    sigma_cap = np.std(training_features)
    
    return -np.sum(np.log(s.norm.pdf(mu_cap,sigma_cap,training_features)))
```

```python
def rayliegh_likelihood(training_features):
    
    training_features = np.array(training_features)
    
    sigma_cap = np.sqrt(np.mean(training_features**2)/2)
    
    return -np.sum(np.log(s.rayleigh.pdf(sigma_cap,training_features)))
```

```python
def determine_distribution(training_features):
    
    if normal_likelihood(training_features) < rayliegh_likelihood(training_features):
        
        return (np.mean(training_features), np.std(training_features))
    
    else:
        
        return np.sqrt(np.mean(training_features**2)/2)
```

```python
from matplotlib import cm

from mpl_toolkits.mplot3d import Axes3D

import scipy.stats as s

import numpy as np
```

```python
def plot_multivariate_normal(params):
    
    X = np.linspace(5,30,50)

    Y = np.linspace(5,30,50)
    
    X,Y = np.meshgrid(X,Y)
    
    data = np.empty(X.shape+(2,))
    
    data[:,:,0] = X

    data[:,:,1] = Y
    
    mean_vec = params[0]
    
    cov_mat = params[1]
    
    P_X_equals_x_and_Y_equals_y = s.multivariate_normal.pdf(data,mean=mean_vec,cov=cov_mat)
    
    fig = plt.figure(figsize=(12,12))

    axes = fig.gca(projection='3d')

    axes.plot_surface(X=X,Y=Y,Z=P_X_equals_x_and_Y_equals_y,rstride=3,
                  cstride=3,linewidth=1,antialiased=True,cmap=cm.viridis)

    axes.contourf(X=X,Y=Y,Z=P_X_equals_x_and_Y_equals_y,zdir='z',cmap=cm.viridis,offset=-0.15)

    axes.set_zlim(-0.15,0.2)

    axes.set_zticks(np.linspace(0,0.2,5))

    axes.view_init(27,-21)

    plt.show()
```

```python
plot_multivariate_normal(B_parameters)
```

```python
plot_multivariate_normal(M_parameters)
```

```python
def P_diagnosis_equals_M_on_features(testing_data_features):
    
    likelihood_m = s.multivariate_normal.pdf(testing_data_features,M_parameters[0],M_parameters[1])
    
    likelihood_b = s.multivariate_normal.pdf(testing_data_features,B_parameters[0],B_parameters[1])
    
    posterior_m = likelihood_m/(likelihood_m + likelihood_b)
    
    return posterior_m > 0.5
```
